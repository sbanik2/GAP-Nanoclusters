```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python

def read_json_file(file_path):
    """
    Reads a JSON file and returns its content.
    """
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

def compute_mae(dft_values, gap_values):
    """
    Computes Mean Absolute Error (MAE) between DFT and GAP values.
    """
    dft_values = np.array(dft_values)
    gap_values = np.array(gap_values)
    mae = np.mean(np.abs(dft_values - gap_values))
    return mae

def extract_energy_and_force_data(E_data, F_data):
    """
    Extracts energy and force data from JSON files for both DFT and GAP.
    
    Returns:
    - dft_energies: List of DFT energies.
    - gap_energies: List of GAP energies.
    - dft_forces: List of DFT forces.
    - gap_forces: List of GAP forces.
    """
    dft_energies = []
    gap_energies = []
    for modality, values in E_data.items():
        dft_energies.extend(values["dft_energy"])
        gap_energies.extend(values["gap_energy"])

    dft_forces = []
    gap_forces = []
    for modality, values in F_data.items():
        dft_forces.extend(values["dft_force"])
        gap_forces.extend(values["gap_force"])

    return dft_energies, gap_energies, dft_forces, gap_forces
```

## Test performance

```python
## Energy and forces
```

```python


# Read energy and force data from JSON files
E_json_file = 'E_pred.json'  # Replace with the actual path to your E.json
F_json_file = 'F_pred.json'  # Replace with the actual path to your F.json

E_data = read_json_file(E_json_file)
F_data = read_json_file(F_json_file)

# Extract energy and force values for plotting
dft_energies, gap_energies, dft_forces, gap_forces = extract_energy_and_force_data(E_data, F_data)

# Compute MAE for energy and force
energy_mae = compute_mae(dft_energies, gap_energies)*1000
force_mae = compute_mae(dft_forces, gap_forces)*1000

# --- Plotting Energy Parity ---
plt.figure(figsize=(4, 4))
plt.scatter(dft_energies, gap_energies, color='b', label='Energy')
plt.plot([min(dft_energies), max(dft_energies)], [min(dft_energies), max(dft_energies)], 'r--')
plt.xlabel('DFT Energy (eV/atom)')
plt.ylabel('GAP Energy (eV/atom)')
plt.title(f'Energy Parity Plot (MAE: {energy_mae:.2f} meV/atom)')
plt.legend()
plt.tight_layout()
plt.show()

# --- Plotting Force Parity ---
plt.figure(figsize=(4, 4))
plt.scatter(dft_forces, gap_forces, color='g', label='Force')
plt.plot([min(dft_forces), max(dft_forces)], [min(dft_forces), max(dft_forces)], 'r--')
plt.xlabel('DFT Force (eV/$\AA$)')
plt.ylabel('GAP Force (eV/$\AA$)')
plt.title(fr'Force Parity Plot (MAE: {force_mae:.2f} meV/$\AA$)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input dictionary (example)
data = read_json_file('lattice_GS.json')

# Extracting polymorph names, DFT, and GAP energies
polymorphs = list(data.keys())
dft_energies = [data[poly]["dft_energy"] for poly in polymorphs]
gap_energies = [data[poly]["gap_energy"] for poly in polymorphs]

# Calculate cohesive energies (energy per atom)
# Assuming the energy values are per unit cell, and considering the number of atoms in the unit cell for each polymorph
# For simplicity, we'll assume the number of atoms per unit cell for each polymorph (a rough approximation).

atoms_per_unit_cell = {"bcc": 2, "hcp": 6, "dc": 1, "fcc": 4}  # These values can be adjusted based on the real unit cell structure

dft_cohesive_energies = [dft / atoms_per_unit_cell[poly] for poly, dft in zip(polymorphs, dft_energies)]
gap_cohesive_energies = [gap / atoms_per_unit_cell[poly] for poly, gap in zip(polymorphs, gap_energies)]

# Plotting the cohesive energies for DFT and GAP
fig, ax = plt.subplots(figsize=(6, 3))

x = np.arange(len(polymorphs))  # The positions for the bars
width = 0.25  # Bar width

rects1 = ax.bar(x - width / 2, dft_cohesive_energies, width, label='DFT', color='b')
rects2 = ax.bar(x + width / 2, gap_cohesive_energies, width, label='GAP', color='g')

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Polymorphs')
ax.set_ylabel('Cohesive Energy (eV/atom)')
ax.set_title('DFT vs GAP Cohesive Energies for Different Polymorphs')
ax.set_xticks(x)
ax.set_xticklabels(polymorphs)
ax.legend(loc=4)

def add_labels(rects, energies, x_offset=0):
    """
    Annotates the bars with their respective energy values.

    Parameters:
    - rects: The bar patches (rectangles).
    - energies: The list of energy values for each bar.
    - x_offset: The horizontal offset to add space between DFT and GAP text labels.
    """
    for rect, energy in zip(rects, energies):
        height = rect.get_height()
        # Horizontal offset (x_offset) is applied here
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2 + x_offset, height-0.2),  # Added x_offset to move text
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adding labels with horizontal gap between DFT and GAP text
add_labels(rects1, dft_cohesive_energies ,x_offset=-0.05)  # DFT annotations without offset

# Apply horizontal offset for GAP annotations to shift them to the right
add_labels(rects2, gap_cohesive_energies, x_offset=0.05)  # Increased horizontal gap for GAP annotations

```

```python

# Extracting polymorph names
polymorphs = list(data.keys())

# Plotting the lattice parameters for each polymorph
for poly in polymorphs:
    dft_lattice_params = data[poly]["lattice_param_dft"]  # DFT lattice parameters [a, b, c]
    gap_lattice_params = data[poly]["lattice_param_gap"]  # GAP lattice parameters [a, b, c]

    # X positions for the bars
    x = np.arange(3)
    width = 0.2  # Bar width

    # Plotting the lattice parameters for DFT and GAP for each polymorph
    fig, ax = plt.subplots(figsize=(4, 3))

    rects1 = ax.bar(x - width / 2, dft_lattice_params, width, label='DFT', color='b')
    rects2 = ax.bar(x + width / 2, gap_lattice_params, width, label='GAP', color='g')

    # Adding labels, title, and custom x-axis tick labels
    ax.set_xlabel('Lattice Parameters (a, b, c)')
    ax.set_ylabel('Lattice Parameter ()')
    ax.set_title(f'{poly} Lattice')
    ax.set_xticks(x)
    ax.set_xticklabels(['a', 'b', 'c'])
    ax.legend(loc='upper left')
    ax.set_ylim(0,max(dft_lattice_params)+5)


    # Adding labels with horizontal gap between DFT and GAP text
    add_labels(rects1, dft_lattice_params,x_offset=-0.08)  # DFT annotations without horizontal offset
    add_labels(rects2, gap_lattice_params, x_offset=0.08)  # GAP annotations with horizontal offset



```
